```python
mdl = SVC(C=30, gamma=10)
mdl.fit(df,Y)
mg.discrete_scatter(df['x1'], df['x2'],Y)
mg.plots.plot_2d_separator(mdl, X)
yp = mdl.predict(df)
accuracy_score(Y, yp)
```

```python
mg.discrete_scatter(df['x1'], df['x2'],Y)
mg.plots.plot_2d_separator(mdl, X)
```

```python
mk = SVC(kernel='poly', degree=2)
mk.fit(df, Y)
mg.discrete_scatter(df['x1'], df['x2'],Y)
mg.plots.plot_2d_separator(mk, X)
yp = mk.predict(df)
accuracy_score(Y, yp)
```

```python
ml = SVC(kernel='linear')
ml.fit(df, Y)
mg.discrete_scatter(df['x1'], df['x2'],Y)
mg.plots.plot_2d_separator(ml, X)
yp = ml.predict(df)
print(yp)
```
