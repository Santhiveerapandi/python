```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snb
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
```

```python
x = np.linspace(23,48,11)
```

```python
y = x + np.random.normal(0,0.3,11)
```

```python
x1 = x + np.random.normal(0,2.7,11)
x2 = x + np.random.normal(0,12.7,11)
x3 = x + np.random.normal(0,20.7,11)
x4 = x + np.random.normal(0,7.7,11)
```

```python
df = pd.DataFrame({'X':x4, 'T':y}).round(2)
```

```python
plt.scatter(x4,y)
plt.show()
```

```python
y[-1] = y[-1] + 400
```

```python
plt.scatter(x4,y)
plt.show()
```

```python
xx = x4.reshape(-1,1)
```

```python
mdl = LinearRegression()
mdl.fit(xx,y)
```

```python
yp = mdl.predict(xx)
```

```python
plt.scatter(x4,y)
plt.plot(x4,yp)
plt.show()
```

```python
mod = Lasso(alpha=190)
mod.fit(xx,y)
yyp = mod.predict(xx)
plt.scatter(xx,y)
plt.plot(xx,yp)
plt.plot(xx,yyp)
plt.show()
```
