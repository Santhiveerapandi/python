### Classification

```python
x,y = make_blobs(centers=2, cluster_std=4.5, random_state=20)
```

```python
X1,X2 = x[:,0],x[:,1]
```

```python
plt.scatter(X1,X2)
```

```python
mg.discrete_scatter(X1,X2,y)
```

```python
model = DecisionTreeClassifier(max_depth=3)
model.fit(x,y)
```

```python
yp = model.predict(x)
```

```python
mg.discrete_scatter(X1,X2,y)
mg.plots.plot_2d_separator(model,x)
```

```python
accuracy_score(y,yp)
```
