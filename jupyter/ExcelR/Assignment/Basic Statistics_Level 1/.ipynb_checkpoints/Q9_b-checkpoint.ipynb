```python
import scipy as s
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('Q9_b.csv')
df
```

```python
# Creating a dataset 
dataset_sp = df['SP']
dataset_wt = df['WT']
```

```python
#Computing skewness of the dataset.
print(s.stats.skew(dataset_sp, axis=0,bias=True))
print(s.stats.skew(dataset_wt, axis=0,bias=True))
```

```python
#Computing kurtosis of the dataset
print(s.stats.kurtosis(dataset_sp, axis=0,bias=True))
print(s.stats.kurtosis(dataset_wt, axis=0,bias=True))
```

```python
ay = s.stats.norm.pdf(dataset_sp, dataset_sp.mean(), dataset_sp.std())
plt.plot(dataset_sp,ay)
```

```python
ay = s.stats.norm.pdf(dataset_wt, dataset_wt.mean(), dataset_wt.std())
plt.plot(dataset_wt,ay)
```
