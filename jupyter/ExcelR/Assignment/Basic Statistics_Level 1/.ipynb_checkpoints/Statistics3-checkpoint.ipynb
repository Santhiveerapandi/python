```python
import scipy.stats as st
```

```python
import numpy as np
import pandas as pd
```

```python
np.linspace(1,10,101)
```

```python
import matplotlib.pyplot as plt
```

```python
ar = np.linspace(1,10,101)
```

```python
ay = st.norm.pdf(ar, ar.mean(), ar.std())
```

```python
plt.plot(ar,ay)
```

```python
cy = st.norm.cdf(ar,ar.mean(), ar.std())
```

```python
plt.plot(ar,ay)
plt.plot(ar,cy)
```

```python
p3 = st.norm.cdf(3,ar.mean(), ar.std())
```

```python
p2 = st.norm.cdf(2,ar.mean(), ar.std())
```

```python
p3 - p2
```
