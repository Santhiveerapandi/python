```python
import scipy as s
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('Q9_a.csv')
df
```

```python
# Creating a dataset 
dataset_speed = df['speed']
dataset_dist = df['dist']
```

```python
#Computing skewness of the dataset.
print(s.stats.skew(dataset_speed, axis=0,bias=True))
print(s.stats.skew(dataset_dist, axis=0,bias=True))
```

```python
#Computing kurtosis of the dataset
print(s.stats.kurtosis(dataset_speed, axis=0,bias=True))
print(s.stats.kurtosis(dataset_dist, axis=0,bias=True))
```

```python
ay = s.stats.norm.pdf(dataset_speed, dataset_speed.mean(), dataset_speed.std())
plt.plot(dataset_speed,ay)
```

```python
ay = s.stats.norm.pdf(dataset_dist, dataset_dist.mean(), dataset_dist.std())
plt.plot(dataset_dist,ay)
```
