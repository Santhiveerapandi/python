```python
Bootstrap plot
Bootstrap plots visually assess a statistics uncertainty, such as mean, median, midrange, etc. 
Bootstrapping involves calculating a statistic by randomly sampling from the same data multiple times,
and then averaging each samples individual result. The resultant values obtained from each random sample are 
then plotted in the form of line and bar charts.
```

```python
from pandas.plotting import bootstrap_plot
bootstrap_plot(df['salary'])
```

# seaborn library

```python

Univariate  x only (contains only one axis of information)
Bivariate  x and y (contains two axis of information)
Trivariate  x, y, z (contains three axis of information)

Scatter Plot. A scatter plot is used to visualize the relationship between two variables. 
Seaborn's scatterplot() function provides a simple way to create scatter plots.

Line Plot. A line plot is used to visualize the trend of a variable over time.
Seaborn's lineplot() function provides a simple way to create line plots.

Histogram. A histogram is used to visualize the distribution of a variable. 
Seaborn's histplot() function provides a simple way to create histograms.

Box Plot. A box plot is used to visualize the distribution of a variable. 
Seaborn's boxplot() function provides a simple way to create box plots.

Violin Plot. A violin plot is similar to a box plot, but provides a more detailed view of the distribution of the data.
Seaborn's violinplot() function provides a simple way to create violin plots.

Heatmap. A heatmap is used to visualize the correlation between different variables.
Seaborn's heatmap() function provides a simple way to create heatmaps.

Pairplot. A pairplot is used to visualize the relationship between multiple variables.
Seaborn's pairplot() function provides a simple way to create pairplots.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
print(sns.get_dataset_names())
```

```python
tips = sns.load_dataset("tips")
print(tips.head())
```

```python
tips['time'].value_counts()
```

```python
tips['total_bill'].corr(tips['tip'])
```

```python
style.use('fivethirtyeight')
# customize the scatter plot
sns.scatterplot(x="total_bill", y="tip" ,hue="time", size="size", sizes=(1,200), data=tips)
# add labels and title
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Relationship between Total Bill and Tip",c='k')
# display the plot
plt.show()
```

```python
element : {"bars", "step", "poly"}
```

```python
sns.histplot(data=tips,x="tip",stat='density',binrange=(1,5),multiple = 'layer',element = 'step',kde=True)
```

```python
# create a FacetGrid for day vs total_bill
g = sns.FacetGrid(tips, col="day")
# plot histogram for total_bill in each day
g.map(sns.histplot, "tip")
```

```python
sns.kdeplot(data=tips, x="total_bill", hue ="day", fill=True, alpha=0.6, linewidth=5)
# Add a title and labels to the plot using Matplotlib
plt.title("Density Plot of Total Bill by Meal Time")
plt.xlabel("Total Bill ($)")
plt.ylabel("Density")
# Show the plot
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", hue="time", data=tips, palette="Set3", linewidth=1.5, fliersize=5)
```

```python
fmri = sns.load_dataset("fmri")
print(fmri.head())
print(fmri['region'].value_counts())
```

```python
# customize the line plot
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", markers=True, dashes=True, data=fmri)
# add labels and title
plt.xlabel("Timepoint")
plt.ylabel("Signal Intensity")
plt.title("Changes in Signal Intensity over Time")
# display the plot
plt.show()
```

```python
titanic = sns.load_dataset("titanic")
print(titanic.head())
```

```python
# customize the bar plot
sns.barplot(x="pclass", y="fare", hue="sex",errorbar = 'ci', palette="bright", data=titanic)
# add labels and title
plt.xlabel("Class")
plt.ylabel("Fare")
plt.title("Average Fare by Class and Gender on the Titanic")
# display the plot
plt.show()
```

```python
iris = sns.load_dataset("iris")
print(iris)
print(iris['species'].value_counts())
```

```python
sns.kdeplot(x="petal_length", data=iris, label='petal_length')
sns.kdeplot(x="sepal_length", data=iris, label='sepal_length')
sns.kdeplot(x="petal_width", data=iris, label='petal_width')
sns.kdeplot(x="sepal_width", data=iris, label='sepal_width')
plt.legend()
```

```python
"point", "stick"
```

```python
sns.violinplot(x="species", y="petal_length", data=iris,inner = 'box')
# display the plot
plt.show()
```

```python
# Create pair plot
sns.pairplot(data=iris,diag_kind ='kde',kind = 'reg')
# Show plot
plt.show()
```

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=iris)
plt.show()
```
