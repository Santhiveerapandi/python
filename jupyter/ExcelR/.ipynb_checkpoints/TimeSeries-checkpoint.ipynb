```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
plot_acf(dx['#Passengers'], lags=50)
plt.show()
```

```python
idx = pd.to_datetime(dx['Month'])
```

```python
dx.set_index(idx, inplace=True)
```

```python
df = dx.drop('Month', axis=1)
```

```python
df.rename(columns={'#Passengers': 'pass'}, inplace=True)
```

```python
df['L2'] = df['pass'].shift(-2)
```

```python
df
```

```python
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
adfuller(df['pass'])[1]
```

```python
d1 = df['pass'] - df['L1']
```

```python
D1 = d1.dropna()
```

```python
adfuller(D1)[1]
```

```python
plt.plot(D1)
```

```python
mod = ARIMA(df['pass'], order=(2,2,14)).fit()
```

```python
yp = mod.predict(start='1958-08-01', end='1962-08-01')
```

```python
plt.plot(df['pass'])
plt.plot(yp)
```

```python
s1 = df['pass'] - df['pass'].shift(-12)
```

```python
S1 = s1.dropna()
```

```python
adfuller(S1)
```

```python
mdl = SARIMAX(df['pass'], order=(2,2,2), seasonal_order=(2,1,2,12)).fit()
```

```python
Syp = mdl.predict(start='1958-08-01', end='1962-08-01')
```

```python
plt.plot(df['pass'])
plt.plot(Syp)
```
