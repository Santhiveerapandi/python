```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(2,10,10)
```

```python
np.random.seed(21)
y = x + 1 + np.random.normal(0,0.5,10)
```

```python
plt.scatter(x,y)
# plt.plot(ar, ar+1)
```

```python
w=0
b=0
plt.scatter(x,y)
yp = w*x + b
plt.plot(x,yp)
for k in range(40):
    w = w + 2*np.mean(x*(y-yp))*0.01
    b = b + 2*np.mean(y-yp)*0.01
    yp = w*x + b
    print(w,b)
    plt.plot(x,yp)
```
