```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(2,10,10)
```

```python
y=x+1+np.random.normal(0,0.5,10)
```

```python
plt.scatter(x,y)
```

```python
w=0
b=0
plt.scatter(x,y)
yp = w*x+b
plt.plot(x,yp)
for k in range(20):
    w= w+2*np.mean(x*(y-yp))*0.01
    b=b+ 2*np.mean(y-yp)*0.01
    yp=w*x+b
    #print(w,b)
    plt.plot(x,yp)
```
